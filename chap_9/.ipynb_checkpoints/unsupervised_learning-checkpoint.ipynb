### 1. Clustering

Group similar instances into clusters.  
Great tool for data analysis, customer segmentation, recommender systems, search engines, semi- supervised learning, dimensionality reduction, anomaly detection, image segmentation ... 

### 2. Anomaly dectection
The object is to learn what a "normal" data looks like, and then use that to detect abnormal instances, such as defective items on production line or a new trend in time series.

### 3. Density estimation
Estimating Probability Density Function (PDF) of the random process that generated the dataset.  
Density estimation is used for Anomaly Detection : instances located in very low density profile regions are likely to be anomalies. Useful for data analysis and visualization.

KMeans initialisation : the algorithm with a complete random initialisation could get stuck on local minima and hence it doesn't find optimal solution. To counter this flaw 2 methods are possible :
 * if you know approximatively where centroids should be, it's possible to init their postion
 * KMeans++ is the algo by default

```python
kmeans = KMeans(n_clusters=4, init=np.array([0,3], [2,3],[10,1], [20,40]), n_init=1)
# n_init will repeat the initialisation n_init time and then take the best solution
# The best solution is the one with the lowest inertia (mean squared distance between point and centroid) 
kmeans.inertia_
```

#### Compute the number of cluster

Silouhette score is a way to find an optimal number of clusters, it computes for all instances (b-a)/max(a,b) (between -1 & 1), where a is the mean intra-cluster (ie the mean distance to all other instances of the cluster) and b is the mean nearest cluster distance (ie the mean distance to the instances of the closest cluster).  
You can compute the silouhette score to have an insight of how good your number of cluster is (when it is optimal it's the closest to +1)

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, kmeans.label_)
```

Silouhette diagramm gives even a better idea of the optimal n_clusters by drawing k knife-shape diagramm.

#### Clustering for Preprocessing

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt 
```

```python
X_digits, y_digits = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

log_reg.score(X_test, y_test)
```

```python
pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=50)),
    ("log_reg", LogisticRegression()),
])

pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test) 
```

```python
param_grid = dict(kmeans__n_clusters=range(2,100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
grid_clf.fit(X_train, y_train)

print(grid_clf.score(X_test, y_test))
print(grid_clf.best_params_)
```

```python
best_k_clusters = grid_clf.best_params_['kmeans__n_clusters']
pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=best_k_clusters)),
    ("log_reg", LogisticRegression()),
])

pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test) 
```

Representative image = image closest to the centroid

```python
k=50
kmeans=KMeans(n_clusters=k)
X_digit_dist = kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digit_dist, axis=0)
X_representative_digit = X_train[representative_digit_idx]
```

```python
plt.gray() 
plt.matshow(X_representative_digit) 
plt.show() 
```

```python
# Now let's manually labelised it (they are supposed to be non labelised)
y_representative_digit = y_train[representative_digit_idx]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_representative_digit, y_representative_digit)
log_reg.score(X_test, y_test)
```

We can propagate the labelisation to the member of the same cluster

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] = y_representative_digit[i]
    
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

To gain accuracy we exclude the digit of the border of the clusters because they are likely to be mislabeled. Here we take only 20% of the most significant digits per cluster.

```python
percentile_closest = 20

X_cluster_dist = X_digit_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_dist = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_dist)
    X_cluster_dist[in_cluster & above_cutoff] = -1
    
partially_propagated = (X_cluster_dist !=-1)
X_partially_propagated = X_train[partially_propagated]
y_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_partially_propagated, y_partially_propagated)
log_reg.score(X_test, y_test)
```

```python
np.mean(y_partially_propagated == y_train[partially_propagated])
```

#### DBSCAN

1. For each instances counts the number of neighbors within eps range
2. If an instance has min_samples neighbors, it's a core instance.
3. All instance in the neighborhood of a core are in the same cluster

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
```

```python
print(dbscan.labels_) # labels associated, index -1 = anomaly, instances doesn't have close neighboors
print(len(dbscan.core_sample_indices_)) # the number of instances with min_samples neighboors
print(dbscan.core_sample_indices_) # the index of instances wich have min_samples neighboors
print(dbscan.components_) # the instances themselves
```

DBSCAN doesn't have a predict method but it's possible to implement one as follow with KNeighorsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.5,0], [0,0.5], [1,-0.1], [2,1]])
print(knn.predict(X_new))
knn.predict_proba(X_new)
```

if an instance is far away from all clusters by default KNeighborsClassifier will class them in one cluster, to avoid that we could compute the distance we could use kneighbors method wich given an instance return a set of instances (n_neighbors instances) that are the k nearest neighbors of the instance and the distances from them

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist>0.2] = -1
y_pred.ravel()
```

```python

def display_image(img, nrows, ncols, figsize):
    # settings
    #nrows, ncols = 5, 4  # array of sub-plots
    #figsize = [6, 8]     # figure size, inches


    # create figure (fig), and array of axes (ax)
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize)

    # plot simple raster image on each sub-plot
    for i, axi in enumerate(ax.flat):
        # i runs from 0 to (nrows*ncols-1)
        # axi is equivalent with ax[rowid][colid]
        #img = face_image_dataset.images[i]
        axi.imshow(img[i], alpha=1)
        # get indices of row/column
        rowid = i // ncols
        colid = i % ncols
        # write row/col indices as axes' title for identification
        axi.set_title("Row:"+str(rowid)+", Col:"+str(colid))

    plt.gray()
    plt.tight_layout(True)

    
display_image(face_image_dataset.images, nrows, ncols, figsize)
plt.show()
```

```python
sss = StratifiedShuffleSplit(test_size=0.2)
sss.get_n_splits(face_image_dataset)

for train_index, test_index in sss.split(face_image_dataset.data,face_image_dataset.target):
    image_train, image_test = face_image_dataset.images[train_index], face_image_dataset.images[test_index]
    X_train, X_test = face_image_dataset.data[train_index], face_image_dataset.data[test_index]
    y_train, y_test = face_image_dataset.target[train_index], face_image_dataset.target[test_index]
```

```python
n_clusters = list(range(37, 43))

for n_cluster in n_clusters:
    
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 20)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.5, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_cluster + 1) * 10])
    
    clusterer = KMeans(n_clusters=n_cluster, random_state=10)
    cluster_labels = clusterer.fit_predict(X_train)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X_train, cluster_labels)
    print("For n_clusters =", n_cluster,
          "The average silhouette_score is :", silhouette_avg)
    sample_silhouette_values = silhouette_samples(X_train, cluster_labels)
    
    y_lower = 10
    for i in range(n_cluster):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_cluster)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_cluster)
    ax2.scatter(X_train[:, 0], X_train[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_cluster),
                 fontsize=14, fontweight='bold')

plt.show()
    
    
```

```python
def plot_precision_recall_vs_threshold(y_train, y_predict):
    precisions, recalls, thresholds = precision_recall_curve(y_train, y_predict)
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.grid()
    #plt.axis([-5, 1, 0, 5])
    plt.xlabel('Threshold')
    plt.legend()

def plot_roc_curve(y_train, y_predict, trace_type="b-", label=None):
    fpr, tpr, thresholds = roc_curve(y_train, y_predict)
    plt.plot(fpr, tpr, trace_type, linewidth=2, label=label)
    plt.plot([0,1], [0,1], 'k--') #Dashed diagonal
    plt.grid("on")
    plt.ylabel('tpr')
    plt.xlabel('fpr')


def display_score(model, X_train, y_train, cv=5):
    #model.fit(X_train, y_train)
    score = cross_val_score(model, X_train, y_train, scoring="accuracy", cv=cv)
    y_predict = cross_val_predict(model, X_train, y_train, cv=cv)
    
    print("Scores : ", score)
    print("Mean : ", score.mean())
    print("Standard deviation :", score.std())
    print("Confusion Matrix : \n", confusion_matrix(y_train, y_predict))
    print("Precision : ", precision_score(y_train, y_predict))
    print("Recall : ",recall_score(y_train, y_predict))
    print("F1_score", f1_score(y_train, y_predict))
    print("ROC Area Under Curve :", roc_auc_score(y_train, y_predict))
    
    plot_precision_recall_vs_threshold(y_train, y_predict)
    plot_roc_curve(y_train, y_predict)
```
